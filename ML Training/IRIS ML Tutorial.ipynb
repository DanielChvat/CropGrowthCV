```python
# Load the Dataset
import pandas
from ucimlrepo import fetch_ucirepo 

iris = fetch_ucirepo(id=53) 
X = iris.data.features 
y = iris.data.targets
X['class'] = y['class']
```

```python
# Sepal Length Histogram
X['sepal length'].hist()
```

```python
# Petal Length Histogram
X['petal length'].hist()
```

```python
# Sepal Width Histogram
X['sepal width'].hist()
```

```python
# Petal Width Histogram
X['petal width'].hist()
```

```python
%matplotlib inline
# Show Petal Length vs Petal Width Scatter Plot for 3 Species
import matplotlib.pyplot as plt
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
X['class'] = y['class']
colors = ['blue', 'green', 'red']
for i in range(3):
    data = X[X['class'] == labels[i]];
    plt.scatter(data['petal length'], data['petal width'], c = colors[i], label = labels[i])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend()
```

```python
# Prepare Data for Training at 75% train and 25% test ratio
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X['class'] = le.fit_transform(X['class'])
x_train, x_test, y_train, y_test = train_test_split(X.drop(columns=['class'], axis='columns'), X['class'], test_size = 0.25)
```

```python
# Train Linear Regression Model and Evaluate accuracy for Test and Train Data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
print('Train Accuracy: ', model.score(x_train, y_train) * 100)
print('Test Accuracy: ', model.score(x_test, y_test) * 100)
```

```python
# Create and Display the Linear Regression Model's Confusion Matrix
from sklearn.metrics import confusion_matrix
import seaborn as sn
y_predict = model.predict(x_test)
cm = confusion_matrix(y_test, y_predict)
plt.figure()
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# Train Decision Tree Model and Evaluate accuracy for Test and Train Data
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
print('Train Accuracy: ', model.score(x_train, y_train) * 100)
print('Test Accuracy: ', model.score(x_test, y_test) * 100)
```

```python
# Create and Display the Decision Tree Model's Confusion Matrix
y_predict = model.predict(x_test)
cm = confusion_matrix(y_test, y_predict)
plt.figure()
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# Train Random Forest Model and Evaluate accuracy for Test and Train Data
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=30)
model.fit(x_train, y_train)
print('Train Accuracy: ', model.score(x_train, y_train) * 100)
print('Test Accuracy: ', model.score(x_test, y_test) * 100)
```

```python
# Create and Display the Random Forest Model's Confusion Matrix
y_predict = model.predict(x_test)
cm = confusion_matrix(y_test, y_predict)
plt.figure()
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
